```python


def find_max_num(lista):
    maximo = lista[0]  # Asumimos que el primer número es el máximo inicialmente
    for numero in lista:
        if numero > maximo:
            maximo = numero
    return maximo

# Probemos la función
numeros = [3, 5, 1, 2, 4]
maximo = find_max_num(numeros)
print("El número máximo es:", maximo)

```

```python
def two_sums(lista, target):
    mydic = {}  # Usamos un diccionario para almacenar los números y sus índices
    for i, num in enumerate(lista):
        complemento = target - num  # Calculamos el complemento necesario para alcanzar el target
        if complemento in mydic:
            return [mydic[complemento], i]  # Si el complemento ya está en el diccionario, retornamos los índices
        mydic[num] = i  # Guardamos el número y su índice en el diccionario
    return []  # Retornamos una lista vacía si no encontramos ninguna pareja

# Probando la función
lista = [2, 7, 11, 15]
target = 9
resultado = two_sums(lista, target)
print(resultado)  # Output: [0, 1]
```

```python
def three_sums(lista, target):
    lista.sort()  # Es más fácil trabajar con una lista ordenada para este tipo de problemas
    for i in range(len(lista) - 2):  # Asegúrate de dejar espacio para dos números más
        if i > 0 and lista[i] == lista[i - 1]:  # Evita duplicados
            continue
        left, right = i + 1, len(lista) - 1
        while left < right:
            suma = lista[i] + lista[left] + lista[right]
            if suma == target:
                return [i, left, right]  # Devuelve los índices de los elementos que suman al target
            elif suma < target:
                left += 1
            else:
                right -= 1
    return []  # Retorna una lista vacía si no se encuentra una solución

# Prueba de la función
lista = [1, 2, -1, -4, 2, -1, 4]
target = 2
resultado = three_sums(lista, target)
print("Índices de los elementos que suman al target:", resultado)
```

```python
import folium
from pyspark.sql import Row

#filtrar los registros que tienen coordenadas validas, no nulas

filtered_df = df.filter(df['Latitude'].isNotNull() & df['Longitude'].isNotNull())

#Convertir el dataframe de spark a una lista de diccionarios(limitado a 100)

restaurant_data = filtered_df.select('DBA','Latitude','Longitude').limit(100).collect()

#Crear un mapa centrado en nueva york

nyc_map = folium.Map(location=[40.7128, -74.0060], zoom_start=11)

#Añadir marcadores para cada restaurante en el mapa

for row in restaurant_data:
  folium.Marker(
      location=[row['Latitude'], row['Longitude']],
                popup=row['DBA'],
  ).add_to(nyc_map)

  nyc_map
```

```python
import matplotlib.pyplot as plt

#agrupar los restaurantes por barrio (BORO) y contar cuantos hay en cada uno

borough_counts = df.groupBy('BORO').count().collect()

#Extraer los nombres de los barrios y las cuentas

boroughs = [row['BORO'] for row in borough_counts]
counts = [row['count'] for row in borough_counts]

#crear el grafico de barras

plt.bar(boroughs, counts, color = 'skyblue')

#personalizar el grafico

plt.xlabel('Barrios')
plt.ylabel('Cantidad de restaurantes')
plt.title('Cantidad de restaurantes por barrio en New York City')
plt.xticks(rotation=45)

plt.show()
```
